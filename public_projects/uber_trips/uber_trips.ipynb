```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/home/pydev/workflow/dt_learn_data_science/public_projects/uber_trips/uber-raw-data-sep14.csv")
data["Date/Time"] = data["Date/Time"].map(pd.to_datetime)
data.head()
```

```python
data["Day"] = data["Date/Time"].apply(lambda x : x.day)
data["WeekDay"] = data["Date/Time"].apply(lambda x : x.weekday())
data["Hour"] = data["Date/Time"].apply(lambda x : x.hour)
data.head()
```

```python
sns.set(rc={'figure.figsize':(12, 10)})
sns.displot(data["Day"], kde=True)
plt.show()
```

```python
sns.displot(data["Hour"],kde=True)
plt.show()
```

```python
sns.displot(data["WeekDay"],kde=True)
plt.show()
```

```python
df = data.groupby(["WeekDay","Hour"]).apply(lambda x: len(x))
df = df.unstack()
sns.heatmap(df, cmap="YlGnBu",annot=False)
```

```python
data.plot(kind='scatter', x='Lon', y='Lat', alpha=0.4, s=data['Day'], label='Uber Trips', figsize=(12,8))
plt.title("Uber Trips Analysis")
plt.legend()
plt.show()
```
