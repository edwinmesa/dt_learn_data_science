# Workers With The Highest Salaries


```python
# Workers With The Highest Salaries

work = pd.read_csv("salaries.csv", sep="\t")
titl = pd.read_csv("title.csv", sep="\t")
work.head(1)
```

```python
titl.head(1)
```

```python
# You have been asked to find the job titles of the highest-paid employees.
# Your output should include the highest-paid title or multiple titles with the same salary.
# 1 Way
work.merge(
    titl, left_on="worker_id", right_on="worker_ref_id", how="inner"
).sort_values(by="salary", ascending=False).head(2)
```

```python
# 2 Way
df_wa_2 = work.merge(titl, left_on="worker_id",
                     right_on="worker_ref_id", how="inner")
df_wa_2[df_wa_2["salary"] == df_wa_2["salary"].max()]["worker_title"]
```

```python
# 3 Way
df_wa_3 = work.merge(titl, left_on="worker_id",
                     right_on="worker_ref_id", how="inner")
df_wa_3["rank"] = df_wa_3.salary.rank(method="dense", ascending=False)
df_wa_3[df_wa_3["rank"] == 1]
```

# Salarys MS Employees


We have a table with employees and their salaries, however, some of the records are old and contain outdated salary information. Find the current salary of each employee assuming that salaries increase each year. Output their id, first name, last name, department ID, and current salary. Order your list by employee ID in ascending order.


```python
ms_empl = pd.read_csv("ms_salary_employee.csv", sep="\t")
ms_empl.head(5)
```

```python
# One Way
# ms_empl = ms_empl[["id", "first_name", "last_name", "department_id", "salary"]]
ms_empl.sort_values("salary", ascending=False).drop_duplicates(
    ["id", "first_name", "last_name", "department_id"]
).sort_index()
```

```python
# Tree Way
ms_empl.groupby(["id", "first_name", "last_name", "department_id"])[
    "salary"
].max().reset_index().sort_values(by="id")
```

# Apple: Count the number of user events performed by MacBookPro users


Count the number of user events performed by MacBookPro users.
Output the result along with the event name.
Sort the result based on the event count in the descending order.


```python
playbook_events = pd.read_csv("playbook_events.csv", sep="\t")
playbook_events.head()
```

```python
# one way
df_macbook = (
    playbook_events[playbook_events["device"] == "macbook pro"]
    .groupby(["event_name"])[["user_id"]]
    .count()
    .reset_index()
)
df_macbook
```

```python
# two way
playbook_events.loc[
    playbook_events.device == "macbook pro", "event_name"
].value_counts().reset_index()
```

# facebook : Popularity of Hack


Meta/Facebook has developed a new programing language called Hack.To measure the popularity of Hack they ran a survey with their employees. The survey included data on previous programing familiarity as well as the number of years of experience, age, gender and most importantly satisfaction with Hack. Due to an error location data was not collected, but your supervisor demands a report showing average popularity of Hack by office location. Luckily the user IDs of employees completing the surveys were stored.
Based on the above, find the average popularity of the Hack per office location.
Output the location along with the average popularity.


```python
facebook_employees = pd.read_csv("facebook_employees.csv", sep="\t")
facebook_hack_survey = pd.read_csv("facebook_hack_survey.csv", sep="\t")
facebook_employees.head()
```

```python
facebook_hack_survey.head()
```

```python
df_facebook = facebook_employees.merge(
    facebook_hack_survey, left_on="id", right_on="employee_id", how="inner"
)
df_facebook.head(3)
```

```python
# One Way
df_facebook.groupby(["location"])[["popularity"]].mean().reset_index()
```

```python
# Two way
facebook_employees.merge(
    facebook_hack_survey, left_on="id", right_on="employee_id", how="inner"
).groupby(["location"])[["popularity"]].mean().reset_index()
```

# linkedlin : Salaries Differences


Write a query that calculates the difference between the highest salaries found in the marketing and engineering departments. Output just the absolute difference in salaries.


```python
linkedlin_employees = pd.read_csv("linkedlin_employees.csv", sep="\t")
linkedlin_deparments = pd.read_csv("linkedlin_departments.csv", sep="\t")
linkedlin_employees.head(2)
```

```python
linkedlin_deparments.head(5)
```

```python
# One Way
linkedlin_salaries = linkedlin_employees.merge(
    linkedlin_deparments, left_on="department_id", right_on="id", how="left"
)

mark = linkedlin_salaries[linkedlin_salaries["department"] == "marketing"]
mark_max_sal = mark["salary"].max()
eng = linkedlin_salaries[linkedlin_salaries["department"] == "engineering"]
eng_max_sal = eng["salary"].max()
diff = abs(mark_max_sal - eng_max_sal)
diff
```

```python
# two Way
linkedlin_salaries = linkedlin_employees.merge(
    linkedlin_deparments, left_on="department_id", right_on="id", how="left"
)

mark = linkedlin_salaries[linkedlin_salaries["department"] == "marketing"]
eng = linkedlin_salaries[linkedlin_salaries["department"] == "engineering"]
abs(mark["salary"].max() - eng["salary"].max())
```

```python
# three way
(
    linkedlin_salaries[linkedlin_salaries["department"]
                       == "marketing"].salary.max()
    - linkedlin_salaries[linkedlin_salaries["department"]
                         == "engineering"].salary.max()
)
```

# Spotify: Find the top 10 ranked songs in 2010


What were the top 10 ranked songs in 2010?
Output the rank, group name, and song name but do not show the same song twice.
Sort the result based on the year_rank in ascending order.


```python
billboard = pd.read_csv("billboard_top_100_year_end.csv", sep="\t")
billboard.head()
```

```python
# One Way
billboard_top_10 = billboard[billboard["year"] == 2010]
billboard_top_10 = billboard_top_10[["year_rank", "group_name", "song_name"]]
billboard_top_10.sort_values(["year_rank"], ascending=True).drop_duplicates(
    ["group_name", "song_name"]
).sort_index().head(5)
```

```python
df = billboard
df[df["year"] == 2010].sort_values(["year_rank"], ascending=True).drop_duplicates(
    subset="song_name"
).head(10)
```
