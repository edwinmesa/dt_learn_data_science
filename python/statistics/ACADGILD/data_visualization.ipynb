```python
# data visualizations
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats     
%matplotlib inline
```

```python
titanic = pd.read_excel('/home/pydev/workflow/dt_learn_data_science/python/statistics/ACADGILD/titanic3.xls')
titanic
```

```python
sns.countplot(x='sex', data=titanic, hue="survived")
```

```python
sns.barplot(x='sex', y='survived', data=titanic)
```

```python
sns.boxplot(x="sex", y="age", data=titanic, palette="rainbow")
```

```python
sns.boxplot(data=titanic, palette="rainbow", orient="h")
```

```python
sns.stripplot(x="pclass", y="fare", data=titanic)
```

```python
plt.figure(figsize=(15,5))
sns.swarmplot(x="pclass", y="fare", data=titanic)
```

```python
plt.figure(figsize=(15,5))
sns.swarmplot(x="pclass", y="fare",hue="sex", data=titanic)
```
