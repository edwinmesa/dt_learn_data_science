```python
def removeOutlier(col):
    sorted(col)
    quant1, quant2 = col.quantile([0.25,0.75])
    IQR            = quant2 - quant1
    lowerRange     = quant1 - (1.5 * IQR)
    upperRange     = quant2 + (1.5 * IQR)
    return lowerRange, upperRange

lowScore, hightScore = removeOutlier(happinessData["Happiness Score"]) 
happinessData["Happiness Score"] = np.where(happinessData["Happiness Score"]>hightScore, hightScore, happinessData["Happiness Score"])
happinessData["Happiness Score"] = np.where(happinessData["Happiness Score"]<lowScore, lowScore, happinessData["Happiness Score"])


print(lowScore)
print(hightScore)

```
