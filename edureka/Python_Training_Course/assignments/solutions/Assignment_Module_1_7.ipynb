1. You are given a data set, which is present in the LMS, containing the number of
   hurricanes occurring in the United States along the coast of the Atlantic. Load the
   data from the dataset into your program and plot a Bar Graph of the data, taking
   the Year as the x-axis and the number of hurricanes occurring as the Y-axis.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv(
    'D:\workflow\dt_learn_data_science\edureka\datasets\Module_6 and Module-7\Hurricanes.csv')

plt.style.use('ggplot')
fig = plt.figure(figsize=(15, 5))
plt.bar(df.Year, df.Hurricanes, color='green', label="Hurricanes", width=.7)
plt.ylabel("Hurricanes")
plt.xlabel("Years")
plt.title("Number of hurricanes occurring in the United States along the coast of the Atlantic")
plt.ylim = 1.0
plt.legend()
plt.show()
```

3. Plot a piechart of the number of models released by every manufacturer, recorded in the data provided. Also mention the name of the manufacturer with the largest releases.


```python
df2 = pd.read_csv(
    'D:\workflow\dt_learn_data_science\edureka\datasets\Module_6 and Module-7\Cars2015.csv')
df2 = df2[['Make']]
# df2['Count'] = df2.groupby('Make')['Make'].transform('count')
# df2.drop_duplicates('Make',inplace=True)
# df2.sort_values(by='Count',ascending=False,inplace=True)
to_dict = dict()
# autopct=lambda p:f'{(p*sum(to_dict.values())/100)+0.5 :.0f}'
plt.figure(figsize=(10, 10))
for make in set(df2['Make']):
    to_dict[make] = df2[df2.Make == make].Make.count()
plt.pie(to_dict.values(), labels=to_dict.keys(),
        autopct=lambda p: f'{(p*sum(to_dict.values())/100)+0.5 :.0f}')
plt.title('Number of cars per make')
plt.show()
```

Case Study: Retail ---- Focus: Visualize the sales data

```python
# 1. Plot total sales per month for the Year 2011. How have the total sales
# increased over the months in the Year 2011? Which month has the lowest
# sales?

df_BigMartSales = pd.read_csv('D:\workflow\dt_learn_data_science\edureka\datasets\Module_6 and Module-7\BigMartSalesData.csv')
df_BigMartSalesMonth = df_BigMartSales.query("Year == 2011").filter(["Month", "Amount"]).groupby(["Month"], as_index=False).sum()
fig = plt.figure(figsize=(15, 5))
plt.ticklabel_format(axis='y',style='sci',scilimits=(1,12))
plt.style.use('ggplot')
plt.plot(df_BigMartSalesMonth["Month"], df_BigMartSalesMonth["Amount"])
plt.xlabel("Month")
plt.ylabel("Amount")
plt.title("Sales 2011 by Month")
plt.show()
# How have the total sales increased over the months in the Year 2011?
    # The increase was from september 2011 to november 2011

# Which month has the lowest sales?
    # November 2011
```

```python
# 2. Plot a Pie Chart for the Year 2011 Country Wise. Which Country contributes
# highest towards sales?
plt.style.use('ggplot')

df_BigMartSalesCountry = df_BigMartSales.query("Year == 2011").filter(
    ["Country", "Amount"]).groupby(["Country"], as_index=False).sum()
plt.pie(df_BigMartSalesCountry["Amount"], labels=df_BigMartSalesCountry["Country"], shadow=True, startangle=90, autopct="%0.1f")
plt.tight_layout()
plt.show()
```

```python
# 3. Plot Scatter Plot for the invoice amounts and see the concentration of amount.
# In which range have most of the invoice amounts concentrated?
# plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12, 8))
df_BigMartInvoices = df_BigMartSales.filter(["InvoiceDate", "Amount"]).groupby(
    "InvoiceDate", as_index=False).sum()
plt.scatter(df_BigMartInvoices["InvoiceDate"], df_BigMartInvoices["Amount"],color='hotpink', alpha=0.7)
plt.show()
# In which range have most of the invoice amounts concentrated?
    # 20000 - 30000
```
