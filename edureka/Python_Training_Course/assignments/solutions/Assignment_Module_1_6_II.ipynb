1. You are given a data set, which is present in the LMS, containing the number of
   hurricanes occurring in the United States along the coast of the Atlantic. Load the
   data from the dataset into your program and plot a Bar Graph of the data, taking
   the Year as the x-axis and the number of hurricanes occurring as the Y-axis.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv(
    'D:\workflow\dt_learn_data_science\edureka\datasets\Module_6 and Module-7\Hurricanes.csv')

plt.style.use('ggplot')
fig = plt.figure(figsize=(15, 5))
plt.bar(df.Year, df.Hurricanes, color='green', label="Hurricanes", width=.7)
plt.ylabel("Hurricanes")
plt.xlabel("Years")
plt.title("Number of hurricanes occurring in the United States along the coast of the Atlantic")
plt.ylim = 1.0
plt.legend()
plt.show()
```

3. Plot a piechart of the number of models released by every manufacturer, recorded in the data provided. Also mention the name of the manufacturer with the largest releases.


```python
df2 = pd.read_csv(
    'D:\workflow\dt_learn_data_science\edureka\datasets\Module_6 and Module-7\Cars2015.csv')
df2 = df2[['Make']]
# df2['Count'] = df2.groupby('Make')['Make'].transform('count')
# df2.drop_duplicates('Make',inplace=True)
# df2.sort_values(by='Count',ascending=False,inplace=True)
to_dict = dict()
# autopct=lambda p:f'{(p*sum(to_dict.values())/100)+0.5 :.0f}'
plt.figure(figsize=(10, 10))
for make in set(df2['Make']):
    to_dict[make] = df2[df2.Make == make].Make.count()
plt.pie(to_dict.values(), labels=to_dict.keys(),
        autopct=lambda p: f'{(p*sum(to_dict.values())/100)+0.5 :.0f}')
plt.title('Number of cars per make')
plt.show()
```

6. Case Study I - Domain  Education -Focus  Data analysis


```python
# 1. Read the three CSV files, which contain the score of the same students in term1
# for each Subject
# 2. Remove the name and ethnicity column (to ensure confidentiality)
# 3. Fill missing score data with zero
# 4. Merge the three files
# 5. Change Sex(M/F) Column to 1/2 for further analysis
# 6. Store the data in a new file  ScoreFinal.csv
df_DSScore      = pd.read_csv('D:\workflow\dt_learn_data_science\edureka\datasets\Module_6 and Module-7\DSScoreTerm1.csv')
df_MathScore    = pd.read_csv('D:\workflow\dt_learn_data_science\edureka\datasets\Module_6 and Module-7\MathScoreTerm1.csv')
df_PhysicScore  = pd.read_csv('D:\workflow\dt_learn_data_science\edureka\datasets\Module_6 and Module-7\PhysicsScoreTerm1.csv')

# 2. Remove the name and ethnicity column (to ensure confidentiality)
del df_DSScore["Ethinicity"]
del df_MathScore["Ethinicity"]
del df_PhysicScore["Ethinicity"]
# 3. Fill missing score data with zero
df_DSScore.fillna(0, inplace=True)
df_MathScore.fillna(0, inplace=True)
df_PhysicScore.fillna(0, inplace=True)
# 4. Merge the three files
tempData      = df_DSScore.merge(df_MathScore, on="ID", suffixes=("_math", "_ds")).merge(df_PhysicScore, on="ID", suffixes=("_phys", "_ds"))
allScoresData = tempData.filter(["ID", "Score_math", "Score_ds", "Score","Age_math","Sex"]).rename(columns={"Score": "Score_physics", "Age_math": "Age"})
allScoresData.head(5)
```

```python
# 5. Change Sex(M/F) Column to 1/2 for further analysis
allScoresData["Sex"] = [1 if sex == "M" else 2 for sex in allScoresData["Sex"]]
allScoresData.head(5)
```

```python
# 6. Store the data in a new file  ScoreFinal.csv
allScoresData.to_csv("ScoreFinal.csv")
```
