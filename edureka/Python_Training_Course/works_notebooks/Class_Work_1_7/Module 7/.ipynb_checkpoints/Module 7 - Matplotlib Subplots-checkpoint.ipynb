```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
ecomm_profit = pd.read_csv('annual_sales.csv')
ecomm_profit
```

```python
import matplotlib.pyplot as plt

# Syntax = plt.subplots(row, cols)
fig, ax = plt.subplots(2,2,figsize=(10,10))

ax[0,0].plot(ecomm_profit['Ecomm'],ecomm_profit['Q1'],'g--o', label='Q1 Profit')
ax[0,0].set_title('Q1 Profit')
ax[0,0].set_xlabel('Ecomm Companies')
ax[0,0].set_ylabel('in Millions')
ax[0,0].grid()

ax[0,1].plot(ecomm_profit['Ecomm'],ecomm_profit['Q2'],'y--*', label='Q2 Profit')
ax[0,1].set_title('Q2 Profit')
ax[0,1].set_xlabel('Ecomm Companies')
ax[0,1].set_ylabel('in Millions')
ax[0,1].grid()

ax[1,0].plot(ecomm_profit['Ecomm'],ecomm_profit['Q3'],'r--^', label='Q3 Profit')
ax[1,0].set_title('Q3 Profit')
ax[1,0].set_xlabel('Ecomm Companies')
ax[1,0].set_ylabel('in Millions')
ax[1,0].grid()

ax[1,1].plot(ecomm_profit['Ecomm'],ecomm_profit['Q4'],'b-->', label='Q4 Profit')
ax[1,1].set_title('Q4 Profit')
ax[1,1].set_xlabel('Ecomm Companies')
ax[1,1].set_ylabel('in Millions')
ax[1,1].grid()

fig.suptitle('Quaterly Profits of Ecommerce Companies')
plt.savefig("Qtr_Profits_Ecomm.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
ecomm_profit = pd.read_csv('annual_sales.csv')

# Syntax = plt.subplots(row, cols)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, figsize=(5,15))

ax1.plot(ecomm_profit['Ecomm'],ecomm_profit['Q1'],'g--o', label='Q1 Profit')
ax1.set_title('Q1 Profit')
ax1.set_xlabel('Ecomm Companies')
ax1.set_ylabel('in Millions')
ax1.grid()


ax2.plot(ecomm_profit['Ecomm'],ecomm_profit['Q2'],'y--o', label='Q2 Profit')
ax2.set_title('Q2 Profit')
ax2.set_xlabel('Ecomm Companies')
ax2.set_ylabel('in Millions')
ax2.grid()


ax3.plot(ecomm_profit['Ecomm'],ecomm_profit['Q3'],'r--o', label='Q3 Profit')
ax3.set_title('Q3 Profit')
ax3.set_xlabel('Ecomm Companies')
ax3.set_ylabel('in Millions')
ax3.grid()


ax4.plot(ecomm_profit['Ecomm'],ecomm_profit['Q4'],'b--o', label='Q4 Profit')
ax4.set_title('Q4 Profit')
ax4.set_xlabel('Ecomm Companies')
ax4.set_ylabel('in Millions')
ax4.grid()


fig.suptitle('Quaterly Profits of Ecommerce Companies')
fig.tight_layout(pad=3)
plt.savefig("Qtr_Profits_Ecomm_1.png")
plt.show()
```
