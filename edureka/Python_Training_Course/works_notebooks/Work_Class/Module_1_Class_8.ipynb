```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ecomm_profit = pd.read_csv('D:\\workflow\\dt_learn_data_science\\edureka\\works_notebooks\\Class_Work_1_7\\Module 7\\annual_sales.csv')
ecomm_profit
```

```python
fig, ax = plt.subplots(2,2, figsize=(10,10))

ax[0,0].plot(ecomm_profit['Ecomm'], ecomm_profit['Q1'], 'g--o', label='Q1 Profit')
ax[0,0].set_title('Q1 Profit')
ax[0,0].set_ylabel('Ecomm Companies')
ax[0,0].set_xlabel('In millons')
ax[0,0].grid()

ax[0,1].plot(ecomm_profit['Ecomm'], ecomm_profit['Q2'], 'y--*', label='Q2 Profit')
ax[0,1].set_title('Q2 Profit')
ax[0,1].set_ylabel('Ecomm Companies')
ax[0,1].set_xlabel('In millons')
ax[0,1].grid()

ax[1,0].plot(ecomm_profit['Ecomm'], ecomm_profit['Q3'], 'r--^', label='Q3 Profit')
ax[1,0].set_title('Q3 Profit')
ax[1,0].set_ylabel('Ecomm Companies')
ax[1,0].set_xlabel('In millons')
ax[1,0].grid()


ax[1,1].plot(ecomm_profit['Ecomm'], ecomm_profit['Q4'], 'y--*', label='Q4 Profit')
ax[1,1].set_title('Q4 Profit')
ax[1,1].set_ylabel('Ecomm Companies')
ax[1,1].set_xlabel('In millons')
ax[1,1].grid()

fig.suptitle('Quartely Profits of Ecomm Companies')
plt.savefig('Qtr_Profits.png')
plt.show()
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4, figsize=(10,15))

ax1.plot(ecomm_profit['Ecomm'], ecomm_profit['Q1'], 'g--o', label='Q1 Profit')
ax1.set_title('Q1 Profit')
ax1.set_ylabel('Ecomm Companies')
ax1.set_xlabel('In millons')
ax1.grid()

ax2.plot(ecomm_profit['Ecomm'], ecomm_profit['Q2'], 'y--*', label='Q2 Profit')
ax2.set_title('Q2 Profit')
ax2.set_ylabel('Ecomm Companies')
ax2.set_xlabel('In millons')
ax2.grid()

ax3.plot(ecomm_profit['Ecomm'], ecomm_profit['Q3'], 'r--^', label='Q3 Profit')
ax3.set_title('Q3 Profit')
ax3.set_ylabel('Ecomm Companies')
ax3.set_xlabel('In millons')
ax3.grid()


ax4.plot(ecomm_profit['Ecomm'], ecomm_profit['Q4'], 'y--*', label='Q4 Profit')
ax4.set_title('Q4 Profit')
ax4.set_ylabel('Ecomm Companies')
ax4.set_xlabel('In millons')
ax4.grid()

fig.suptitle('Quartely Profits of Ecomm Companies')
fig.tight_layout(pad=3)
plt.savefig('Qtr_Profits_2.png')
plt.show()
```

```python
df = pd.read_csv('D:\workflow\dt_learn_data_science\edureka\works_notebooks\Class_Work_1_7\Module 7\cereal.csv')
df = df.iloc[-15:]
df
```

```python
grid = plt.GridSpec(3,3)
fig = plt.figure(figsize=(10,10))
fig.add_subplot(grid[0,:2])
plt.title('Potassion')
plt.plot(df.potass)
fig.add_subplot(grid[1,:2])
plt.title('Sodium')
plt.plot(df.sodium)
fig.add_subplot(grid[2,:])
plt.title("Calories vs Potassion")
plt.scatter(df.calories,df.potass)
fig.add_subplot(grid[:2,2])
plt.barh(df.index, df.rating)
plt.yticks(df.index)
plt.title('Ratings')
plt.show()

```
