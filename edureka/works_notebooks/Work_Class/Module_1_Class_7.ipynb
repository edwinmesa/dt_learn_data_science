```python
plt.style.use('tableau-colorblind10')
x = [10, 20, 30]
y = [100, 175, 225]
plt.plot(x, y, marker='^')

plt.title('Speed (km/h) v/s Calories burnt on Sunday')
plt.xlabel('Speed (km/h)')
plt.ylabel('Calories Burnt')
plt.show()
```

```python
plt.style.use('ggplot')

plt.plot([10, 20, 30], [100, 175, 225], marker='o', label='Sunday')
plt.plot([10, 20, 30], [120, 200, 250], marker='*', label='Monday')
plt.plot([10, 20, 30], [90, 160, 200], marker='^', label='Tuesday')

plt.title('Speed (km/h) v/s Calories burnt', fontsize=10)
plt.xlabel('Speed (km/h)')
plt.ylabel('Calories burnt')

plt.xticks(np.arange(0, 40, 5))
plt.yticks(np.arange(75, 250, 25))

plt.legend(loc='upper left')
plt.show()
```

```python
plt.style.use('ggplot')

plt.plot([10, 20, 30], [100, 175, 225], 'g', marker='o', label='Sunday')
plt.plot([10, 20, 30], [120, 200, 250], 'y', marker='*', label='Monday')
plt.plot([10, 20, 30], [90, 160, 200], 'r', marker='^', label='Tuesday')

plt.title('Speed (km/h) v/s Calories burnt', fontsize=10)
plt.xlabel('Speed (km/h)')
plt.ylabel('Calories burnt')

plt.xticks(np.arange(0, 40, 5))
plt.yticks(np.arange(75, 250, 25))

plt.legend(loc='upper left')
plt.show()
```

```python
plt.style.use('ggplot')

plt.plot([10, 20, 30], [100, 175, 225], 'g',
         linestyle='--', marker='o', label='Sunday')
plt.plot([10, 20, 30], [120, 200, 250], 'y',
         linestyle='-', marker='*', label='Monday')
plt.plot([10, 20, 30], [90, 160, 200], 'r',
         linestyle=':', marker='^', label='Tuesday')

plt.title('Speed (km/h) v/s Calories burnt', fontsize=10)
plt.xlabel('Speed (km/h)')
plt.ylabel('Calories burnt')

plt.xticks(np.arange(0, 40, 5))
plt.yticks(np.arange(75, 250, 25))

plt.legend(loc='upper left')
plt.show()
```

```python
data_dict = {'CSE': 33, 'ECE': 98, 'GHT': 58, 'NJU': 33}
courses = list(data_dict.keys())
values = list(data_dict.values())

fig = plt.figure(figsize=(5, 5))

plt.bar(courses, values, color="blue", width=0.4)
plt.xlabel("Courses Offered")
plt.ylabel("Students")
plt.title("Student enrolled in different courses")
plt.show()
```
