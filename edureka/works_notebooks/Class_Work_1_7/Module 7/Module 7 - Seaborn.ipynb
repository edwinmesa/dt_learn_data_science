### Relating variables with scatter plots

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="tip", data=tips)
```

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="size", hue='sex', data=tips)
```

```python
tips['sex'].value_counts()
```

```python
'''
While the points are plotted in two dimensions, another dimension can be added to the plot by coloring the points according to a third variable. 
In seaborn, this is referred to as using a “hue semantic”, because the color of the point gains meaning:
'''
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
tips['smoker'].value_counts()
```

```python
tips['sex'].value_counts()
```

```python
tips['day'].value_counts()
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style = "time", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", kind="line", data=tips)
```

### Categorical scatterplots

```python
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips)
```

```python
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], data=tips)
```

### Pairplots

```python
import seaborn as sns
sns.set_theme(style="ticks")

tips = sns.load_dataset("tips")
g = sns.pairplot(tips, hue="smoker")
g.fig.set_size_inches(10,10)
```

### Distribution Plot --> https://seaborn.pydata.org/tutorial/distributions.html

```python
sns.displot(tips, x="total_bill", hue="sex", kind="kde", fill = True)
```

```python
sns.displot(tips, x="total_bill", hue="sex", element="step")
```

```python
# ci = confidence interval
bar_plot = sns.barplot(x="total_bill", y="sex", data = tips, ci = 0)
bar_plot.set_xlim(0, 22)
```

```python
tips.to_csv("tips.csv")
```

```python
fig = plt.figure(figsize=(7,7))
ax = sns.countplot(x="sex", data = tips)
ax.set_ylim(0, 170)
```

```python
sns.boxplot(x="total_bill", y="sex", data = tips)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = pd.read_csv("StudentsPerformance.csv")
print(data.shape)
data.head()
```

```python
# Correlation is a statistical measure that expresses the extent to which two variables are linearly related
print(data.corr())
sns.heatmap(data.corr(), cmap="YlGnBu", annot=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_csv('2016.csv')
sns.set(style = "darkgrid")

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection = '3d')

x = df['Happiness Score']
y = df['Economy (GDP per Capita)']
z = df['Health (Life Expectancy)']

ax.set_xlabel("Happiness")
ax.set_ylabel("Economy")
ax.set_zlabel("Health")

ax.scatter3D(x, y, z)

plt.show()
```
