## Bar Chart

```python
'''
Why do we need a bar graph? 
A bar graph uses bars to compare data among different categories. 
It is well suited when you want to measure the changes over a period of time. 
It can be represented horizontally or vertically. 
Also, the important thing to keep in mind is that longer the bar, greater is the value.
'''
```

```python
# bar chart to check which month rainfall correlates with annual rainfall

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.style.use('ggplot')
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "green")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Dataset generation - CSE : Computer Sc Engg, ECE : Eletronics Engg, EEE : Electrical Engg
data_dict = {'CSE':33, 'ECE':28, 'EEE':30}

courses = list(data_dict.keys()) # x values
values = list(data_dict.values()) # y values

fig = plt.figure(figsize = (5, 5)) # size of the graph

#  Bar plot
plt.bar(courses, values, color ='blue', width = 0.5)
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
# Stacked bar chart
import matplotlib.pyplot as plt
  
# create data
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]
  
# plot bars in stack manner
plt.style.use('seaborn')
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.show()
```

## Pie Chart

```python
'''
Pie charts visualize absolute and relative frequencies. A pie chart is a circle partitioned into segments where each of 
the segments represents a category. A pie chart is drawn using pie() function considering single list as an argument.
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
fig = plt.figure(figsize = (7,7))
plt.pie(y, labels = mylabels)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
fig = plt.figure(figsize = (5,5))
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
# Colors
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["pineapple", "Bananas", "Cherries", "Dates"]
mycolors = ["yellow", "hotpink", "b", "#4CAF50"]
myexplode = [0.1, 0.1, 0, 0]

fig = plt.figure(figsize = (7, 7)) # size of the graph
plt.pie(y, labels = mylabels, colors = mycolors, explode = myexplode, shadow = True)
plt.show() 
```

```python
# Pie Values
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["lightblue", "yellow", "green", "sienna"]

plt.figure(figsize = (10,10))
plt.pie(y, labels = mylabels,  colors = mycolors, autopct="%0.1f", 
        textprops={'family':'sans-serif', 'fontsize':'x-large', 
             'fontstyle':'italic', 'fontweight':'heavy'})
plt.legend(title = "Four Fruits:", loc = 'upper right')
plt.title('Fruits',fontweight='bold')
plt.show() 
```

## Scatter Plot

```python
'''
Usually we need scatter plots in order to compare variables, for example, how much one variable is affected by 
another variable to build a relation out of it. The data is displayed as a collection of points, each having the 
value of one variable which determines the position on the horizontal axis and the value of other variable determines 
the position on the vertical axis.
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,  7, 8, 7, 2, 17,  2, 9, 4,11,12, 9, 6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.style.use('seaborn')
plt.scatter(x, y)
plt.show()
```

## Histogram

```python
'''
Histograms are used to show a distribution whereas a bar chart is used to compare different entities. 
Histograms are useful when you have arrays or a very long list. Lets consider an example where I have 
to plot the age of population with respect to bin. Now, bin refers to the range of values that are divided 
into series of intervals. Bins are usually created of the same size. In the below code, I have created the 
bins in the interval of 10 which means the first bin contains elements from 0 to 9, then 10 to 19 and so on.
'''
```

```python
# https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.hist.html
# plt.hist(population_age, bins, histtype ='bar', rwidth=0.1)
# bin width = ( 120  2 ) / 12  = 9.8
# 30 values 

import matplotlib.pyplot as plt
population_age = [22,55,62,45,21,22,34,42,42,4,2,102,95,85,55,110,120,70,65,55,111,115,80,75,65,54,44,43,42,48]
plt.hist(population_age, bins=5, edgecolor="yellow", color="green")

plt.xlabel('age groups')
plt.ylabel('Number of people')
plt.title('Histogram')
plt.show()
```

## Box Plot

```python
'''
A box and whisker plotalso called a box plotdisplays the five-number summary of a set of data. 
The five-number summary is the minimum, first quartile, median, third quartile, and maximum.
'''
```

```python
'''
A sample of 10 boxes of raisins has these weights (in grams):
25, 28, 29, 29, 30, 34, 35, 35, 37, 38
'''
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.boxplot.html

import matplotlib.pyplot as plt
import numpy as np
rasins_wt = [25, 28, 29, 29, 30, 34, 35, 35, 37, 38]

plt.boxplot(rasins_wt, vert=False)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xticks(np.arange(22,42,2))

plt.show()
```

```python
import numpy
from matplotlib import pyplot

x = numpy.arange(10) # 0 - 9
y = numpy.array([5,3,4,2,7,5,4,6,3,2])

fig = pyplot.figure()

ax = fig.add_subplot(111)
ax.set_ylim(0,10)

pyplot.plot(x,y)

for i,j in zip(x,y):
    ax.annotate(str(j), xy=(i,j))
    print(i, j)
    
pyplot.show()
```
