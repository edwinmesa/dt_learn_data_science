##Demo 1: Using Web Maps to Find the Location of Rental Bike

Here we will find out all the location of rental bikes in Dublin, Ireland.For this problem statement,we use the dataset of Dublin to find the location of all the rental bikes based on their latitude and longitude.

```python
#import all the necessary libraries
import pandas as pd
import folium
```

```python
#read the dataset and display the first five rows
df = pd.read_csv("dublin.csv")
df.head()
```

```python
#for the sake of convenience we will be using three features - latitude, longitude, and name
df1 = df[["Latitude", "Longitude", "Name"]]
```

```python
#The latitude and longitude are passed to the location parameter of Map.
map = folium.Map(location=[df1.Latitude.mean(), df1.Longitude.mean()], zoom_start=14, control_scale=True)
```

```python
map
```

```python
#The markers are pointed on the location where rental bikes can be found.The resulting map is stored in a html file.
m=folium.Map()
for index, location in df1.iterrows():
    marker = folium.Marker([location["Latitude"], location["Longitude"]], popup=location["Name"],icon=folium.Icon(color="green"))
    marker.add_to(m)
m.save("map1.html")
m
```

Conclusion: When we download and open the map1.html file we can see all the locations in Dublin where rental locations of bike can be found.
