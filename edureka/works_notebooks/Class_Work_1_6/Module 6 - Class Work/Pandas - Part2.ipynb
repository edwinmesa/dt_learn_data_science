```python
# Single code to remove all errors...
import pandas as pd
import math
df = pd.read_csv('data.csv')
##########################################

x = df["Calories"].mean().round(1)
print(x)
df["Calories"].fillna(x, inplace = True)
#
y = df["Pulse"].mean().round(1)
format(math.ceil(y),".1f")
print(math.ceil(y))
df["Pulse"].fillna(math.ceil(y), inplace = True)
###############################################

import datetime
now = datetime.datetime.now()
date_time = now.strftime("%d-%m-%Y")
print("date and time:",date_time)
df.at[27,'Date'] = date_time
##############################################

df.at[7,'Duration'] = 45
##############################################
df.drop_duplicates(inplace = True)
df
```

```python
### Adding a new column
```

```python
df['Total'] = df['Pulse'] + df['Maxpulse'] + df['Calories']
df
```

```python
### Revoming a column - use function drop
```

```python
df = df.drop(['Total'], axis='columns')
df
```

```python
### Removing rows
```

```python
df = df.drop(df.index[20:])
df
```
