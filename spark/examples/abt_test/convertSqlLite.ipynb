```python
import pandas as pd
import numpy as np
from pyarrow.parquet import ParquetDataset
import os
import glob
import pyarrow.parquet as pq
from sqlalchemy import create_engine
import sqlite3 as sq

# engine = create_engine('sqlite://', echo=False)
table_name = "tbl_land_distance" # table and file name
conn = sq.connect('{}.sqlite'.format(table_name)) # creates file
```

```python
all_months_data = pd.DataFrame()
files_csv = [f for f in os.listdir("./distance") if f.endswith('.parquet')]
for file in files_csv:
    df = pd.read_parquet("./distance/"+file, engine='pyarrow')
    all_months_data = pd.concat([all_months_data, df])
all_months_data.to_sql(table_name, conn, if_exists='replace', index=False) # writes to file
# all_months_data.to_sql('users', if_exists='replace',con=engine)
conn.close()
# all_months_data.to_csv('./Output/all_data.csv', index=False) 
```
